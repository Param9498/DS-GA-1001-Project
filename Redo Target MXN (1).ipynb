```python
from azureml import Workspace
import math
import matplotlib.pylab as plt

from datetime import datetime as dt

#adjust to each CCY
ws = Workspace()
ds = ws.datasets['USDMXN Input Data.csv']
df = ds.to_dataframe()

ds2 = ws.datasets['USDMXN Trends.csv']
df2 = ds2.to_dataframe()
df2 = df2.drop(['Day Start','SPX Day Start'],axis = 1)

cols_to_use = df2.columns.difference(df.columns)
df2 = df2[cols_to_use]

df = df.reset_index()
df2 = df2.reset_index()

data = df.merge(df2,left_on = 'index', right_on = 'index')
data
```

```python
import numpy as np
import pandas as pd
import seaborn as sn

data['Target New'] = np.where(data['Abs Max Trend Sz'] > .0029,1,0)
data['Target Dif'] = abs(data['Target New'] - data['Target'])
sum(data['Target Dif'])
```

```python
sum(data['Target New'])
```

```python
data['Target'] = data['Target New']
```

```python
data['RR Divergence'] = data['Previous Day RR'] - data['Previous Day RR MAVG']
```

```python
USDMXN_Input_1 = data[['Trend Start Time', 'Daily Trend #', 'Previous Day DXY Trend', 'RR Divergence', 'Previous Day SPY Trend', 
                     'Trend Start', 'Trend Type', 'USDMXNV1M Tick', 'USDMXNV1Y Tick', 'V1M Change', 'V1Y Change', 'SPX Change',
                     'MXEF Tick', 'MXEF CCY Tick', 'BBDXY Tick', 'USDMXN RSI', 'USDMXN MACD', 'USDMXN ROC 4', 'USDMXN ROC 10', 
                     'MXEF Change', 'MXEF CCY Change', 'BBDXY Change', 'Target']]
USDMXN_Input_1
```
